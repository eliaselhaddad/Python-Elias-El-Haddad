```python
import matplotlib.pyplot as plt
import numpy as np

# array from -10 to 10 with 100 numbers equally spaced between
x = np.linspace(-10, 10, 100)
f = lambda x: np.cos(x)
```

## MATLAB style interface

```python
plt.plot(x, f(x), 'o--')
```

```python
x = np.linspace(-10, 10, 200)

plt.figure(figsize=(8, 4), dpi = 200)
plt.plot(x, f(x))
plt.title("cos(x)")
plt.xlabel("x")
plt.ylabel("y")
# plt.box(False)
```
